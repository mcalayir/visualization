```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
from PIL import Image   # Open images
import requests         # Get images from URLs
from io import BytesIO  # Store images
```

```python
url = "https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png"

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "image/*,*/*;q=0.8",
    "Referer": "https://en.wikipedia.org/"
}
```

```python
response = requests.get(url, headers = headers)
image_file = BytesIO(response.content)
image = Image.open(image_file)
image
```

```python
x=np.array(['Luffy','Zoro','Nami','Usopp','Sanji'])
y1=np.array([110,180,240,99,220])
y2=np.array([170,100,90,120,50])
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax_image = fig.add_axes([0.1, # x coordinate 
0.11, # y coordinate 
0.15, # image width
0.35]) # image height
ax_image.imshow(image)
ax_image.axis('off')

plt.savefig('/Users/john/Desktop/DSI/visualization/02_activities/assignments/participation/image1',dpi=300)
plt.show()
```
